```python
# Install Pandas Data Reader
# !pip install pandas-datareader
```

```python
import pandas_datareader as pdr
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pdr.get_data_yahoo("TSLA")
```

```python
df
```

```python
df.plot()
```

```python
df["High"].plot(figsize=(15,5))
```

```python
# xlimt and ylimt (Plot with in a range)
df["High"].plot(xlim=["2020-01-01","2022-09-01"],ylim=[0,500],figsize=(15,5))
```

```python
# xlimt and ylimt (Plot with in a range) and color
df["High"].plot(xlim=["2020-01-01","2022-09-01"],ylim=[0,500],figsize=(15,5),c="r",ls="--")
```

```python
# Values of particular dates
df.loc["2020-01-01":"2020-09-01"]
```

```python
# Index of particular dates
df.loc["2020-01-01":"2020-09-01"].index
```

```python
index=df.loc["2020-01-01":"2020-09-01"].index
share_open=df.loc["2020-01-01":"2020-09-01"]["Open"]
```

```python
index
```

```python
share_open
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
# autofmt will format the labels in x axis (prevent overlapping)
figure.autofmt_xdate()
axis.plot(index,share_open)
```
