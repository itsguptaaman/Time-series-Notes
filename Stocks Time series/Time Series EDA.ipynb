#### Time Resampling
for more refrence https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```python
df
```

```python
# Minimum share price throught the years
```

```python
# Rule A means Year and frequency 
df.resample(rule="A").min()
```

```python
# Maximum share price throught the years
df.resample(rule="A").max()
```

```python
df.resample(rule="A").max()["Open"].plot()
```

```python
# QS Quaterly Sample
df.resample(rule="QS").max()["Open"].plot()
```

```python
#  QS Bussiness End Frequency
df.resample(rule="BA").max()["High"].plot()
```

```python
df.resample(rule="BA").max()
```

```python
# Bussiness Quaters
df.resample(rule="BQS").max()
```

```python
# Graph Plotimg
```

```python
df.Open.resample(rule="A").mean().plot(kind="bar")
```

```python
df.Open.resample(rule="M").max().plot(kind="bar",figsize=(15,6))
```

```python
# Mean Average(Rolling Window)
```

```python
df["High"].rolling(10).mean().head(30)
```

```python
df["Open:30 days rolling"]=df["Open"].rolling(30).mean()
```

```python
df.head(35)
```

```python
df[["Open","Open:30 days rolling"]].plot(figsize=(12,5))
```
